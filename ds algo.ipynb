What is Big 0?

Big O Notation is used to show how efficient an algorithm or function is, in relative to its input size.

To measure the efficiency of an algorithm, we need to consider two things:

â€¢ Time Complexity: How much time does it take to run completely?

Space Complexity: How much extra space does it require in the process?

## Binary Search Algorithm (iteratively)

```python
#Sorted List
def BinarySearch(a, key):
    low=0
    high=len(a)-1
    while low<=high:
        mid = (low+high)//2
        if key==a[mid]:
            return True
        elif key<a[mid]:
            high=mid-1
        else:
            low=mid+1
            return False
a=[15, 21, 47, 84, 96]
found=BinarySearch(a, 47)
print(found)
```

## Binary Search Algorithm (recursively)

```python
def BinarySearch(a, key, low, high):
    if high >= low:
        mid = low + (high - low)//2
        if key==a[mid]:
            return mid
        elif key<a[mid]:
            return BinarySearch(a, key, low, mid-1)
        else:
            return BinarySearch(a, key, mid+1, high)
    else:
        return "not present"
a=[15, 21, 47, 84, 96]
key=int(input())
print(BinarySearch(a, key, 0, len(a)-1))
```

## Selection Sort

```python
# Minimum Position
def SelectionSort(List):
    for i in range (len(List)):
        MinPos=i
        for j in range(i+1, len(List)):
            if List [MinPos]>List[j]:
                MinPos=j
            List[i], List[MinPos]=List[MinPos], List[i]            
    return List
List=[5, 2, 1, 9, 0, 4,6]
print(SelectionSort(List))
```

```python
# Counting sort in Python programming

def countingSort(array):
    size = len(array)
    output = [0] * size
    count = [0] * 10

    for i in range(0, size):
        count[array[i]] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    i = size - 1
    while i >= 0:
        output[count[array[i]] - 1] = array[i]
        count[array[i]] -= 1
        i -= 1

    for i in range(0, size):
        array[i] = output[i]


data = [4, 2, 2, 8, 3, 3, 1]
countingSort(data)
print("Sorted Array in Ascending Order: ")
print(data)
```

## Insertion Sort

```python
def InsertionSort(a):
    for i in range(len(a)):
        value=a[i]
        pos=i-1
        while pos>=0 and a[pos] > value:
            a[pos]=a[pos+1]
            pos=pos-1
            a[pos+1]=value
data=[15, 21, 47, 84, 96]
found=InsertionSort(data)
print(data)
```

## Bubble Sort

```python
def BubbleSort(a):
    for i in range (len(a)-1, 0, -1):
        for j in range (i):
            if a[j] > a[j+1]:
                a[j], a[j+1]=a[j+1], a[j]
data=[15, 21, 47, 84, 74,  96]
found=BubbleSort(data)
print(data)
```

## Quick Sort

```python
def QuickSort(a, low, high):
    if low < high:
        p = partition(a, low, high)
        QuickSort(a, low, p-1)
        QuickSort(a, p+1, high)
        
def partition(a, low, high):
    i=low-1
    pivot=a[high]
    for j in range(low, high):
        if a[j]<=pivot:
            i=i+1
        a[i], a[j]=a[j], a[i]
        
    a[i+1], a[high]= a[high], a[i+1]
    return i+1
a=[84, 21, 96, 15, 47]
QuickSort(a, 0, len(a)-1)
print(a)
```

# Reverse a Linked List

```python


# Node class
class Node:

    # Init method to initialize the node object
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:

    # Init method to initialize head
    def __init__(self):
        self.head = None

    # Reverse method to reverse the linked list
    def reverse(self):
        prev = None
        current = self.head
        while (current is not None):
            next = current.next
            current.next = prev
            prev = current
            current = next
        self.head = prev

    # Push to insert a new node at the beginning
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    # Utility function to print the linked LinkedList
    def display(self):
        temp = self.head
        while (temp):
            print(temp.data)
            temp=temp.next


llist = LinkedList()
llist.push(20)
llist.push(40)
llist.push(60)
llist.push(80)

print("Given Linked List")
llist.display()
llist.reverse()
print("\nReversed Linked List")
llist.display()
```

### Circular Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data 
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None 
        # Function to insert in front(prepend)
    def prepend(self, data):
        new_node = Node(data)
        temp = self.head 
        new_node.next = self.head

        if not self.head:
            new_node.next = new_node
        else:
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
        self.head = new_node
        # Function to insert at the end(append)
    def append(self, data):
        if not self.head:
            self.head = Node(data)
            self.head.next = self.head
        else:
            new_node = Node(data)
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head
        # Function to print
    def print_list(self):
        temp = self.head 

        while temp:
            print(temp.data, end=" ")
            temp = temp.next
            if temp == self.head:
                break
        # Function to delete
    def remove(self, key):
        if self.head.data == key:
            temp = self.head 
            while temp.next != self.head:
                temp = temp.next 
            temp.next = self.head.next
            self.head = self.head.next
        else:
            temp = self.head 
            prev = None 
            while temp.next != self.head:
                prev = temp 
                temp = temp.next
                if temp.data == key:
                    prev.next = temp.next 
                    temp = temp.next


cllist = CircularLinkedList()
cllist.append("A")
cllist.append("B")
cllist.append("C")
cllist.append("D")

cllist.remove("A")
cllist.remove("C")
cllist.print_list()
```

### Doubly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        # Function to insert in end(append)
    def append(self, data):
        if self.head is None:
            new_node = Node(data)
            new_node.prev = None
            self.head = new_node
        else:
            new_node = Node(data)
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp
            new_node.next = None
        # Function to insert in front(prepend)
    def prepend(self, data):
        if self.head is None:
            new_node = Node(data)
            new_node.prev = None 
            self.head = new_node
        else:
            new_node = Node(data)
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
            new_node.prev = None
        # Function to print
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
    def add_after_node(self, key, data):
        temp = self.head
        while temp:
            if temp.next is None and temp.data == key:
                self.append(data)
                return 
            elif temp.data == key:
                new_node = Node(data)
                nxt = temp.next
                temp.next = new_node
                new_node.next = nxt
                new_node.prev = temp
                nxt.prev = new_node
            temp = temp.next

    def add_before_node(self, key, data):
        temp = self.head
        while temp:
            if temp.prev is None and temp.data == key:
                self.prepend(data)
                return
            elif temp.data == key:
                new_node = Node(data)
                prev = temp.prev
                prev.next = new_node
                temp.prev = new_node
                new_node.next = temp
                new_node.prev = prev
            temp = temp.next
        # Function to delete
    def delete(self, key):
        temp = self.head
        while temp:
            if temp.data == key and temp == self.head:
                # Case 1:
                if not temp.next:
                    temp = None 
                    self.head = None
                    return

                # Case 2:
                else:
                    nxt = temp.next
                    temp.next = None 
                    nxt.prev = None
                    temp = None
                    self.head = nxt
                    return 

            elif temp.data == key:
                # Case 3:
                if temp.next:
                    nxt = temp.next 
                    prev = temp.prev
                    prev.next = nxt
                    nxt.prev = prev
                    temp.next = None 
                    temp.prev = None
                    temp = None
                    return

                # Case 4:
                else:
                    prev = temp.prev 
                    prev.next = None 
                    temp.prev = None 
                    temp = None 
                    return 
            temp = temp.next


dllist = DoublyLinkedList()
dllist.prepend(0)
dllist.append(1)
dllist.append(2)
dllist.append(3)
dllist.append(4)
dllist.prepend(5)
dllist.delete(3)
dllist.add_after_node(2, 5)

dllist.print_list()
```

# Reverse a doubly linked list using recursion

```python
# Node of double Linked list
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None

# getnode method for new node
def getNode(data):
	new_node = Node(data)
	new_node.data = data
	new_node.next = new_node.prev = None
	return new_node

# push to insert node at the beginning of the Doubly Linked List
def push(head_ref, new_node):

	# since we are adding at the beginning, prev is always None
	new_node.prev = None

	# link the old list off the new node
	new_node.next = head_ref

	# change prev of head node to new node
	if (head_ref != None):
		head_ref.prev = new_node

	# move the head to point to the new node
	head_ref = new_node
	return head_ref

# Reverse method to reverse a doubly linked list
def Reverse(node):

	# If empty list, return
	if not node:
		return None

	# Otherwise, swap the next and prev
	temp = node.next
	node.next = node.prev
	node.prev = temp

	# If the prev is now None, the list has been fully reversed
	if not node.prev:
		return node

	# Otherwise, keep going
	return Reverse(node.prev)

# Function to print nodes in a given doubly linked list
def display(head):
	while (head != None) :
		print(head.data, end = " ")
		head = head.next

# Start with the empty list
head = None

# Create doubly linked:
head = push(head, getNode(10))
head = push(head, getNode(20))
head = push(head, getNode(30))
head = push(head, getNode(40))
print("Original list: ", end = "")
display(head)

# Reverse doubly linked list
head = Reverse(head)
print("\nReversed list: ", end = "")
display(head)


```

```python


# A Node class to create as many node objects to store queue entry...
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None

# A Queue class, front here stores the front node while rear stores the last node of linked list....
class Queue:

    def __init__(self):
        self.front = self.rear = None

    def isEmpty(self):
        return self.front == None

    # We can add items in the queue using the Enqueue method....
    def Enqueue(self, item):
        temp = Node(item)

        if self.rear == None:
            self.front = self.rear = temp
            return
        self.rear.next = temp
        self.rear = temp

    # We can remove items from the queue using the Dequeue method....
    def Dequeue(self):

        if self.isEmpty():
            return
        temp = self.front
        self.front = temp.next

        if (self.front == None):
            self.rear = None




q = Queue()
q.Enqueue(5)
q.Enqueue(10)
q.Dequeue()
q.Dequeue()
q.Enqueue(15)
q.Enqueue(20)
q.Enqueue(25)
print("Queue Front " + str(q.front.data))
print("Queue Rear " + str(q.rear.data))
```
