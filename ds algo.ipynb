```python
What is Big 0?

Big O Notation is used to show how efficient an algorithm or function is, in relative to its input size.

To measure the efficiency of an algorithm, we need to consider two things:

â€¢ Time Complexity: How much time does it take to run completely?

Space Complexity: How much extra space does it require in the process?

Big-o captures the upper bound to show how much time or memory an algorithm would require in the worst case scenario as the input size grows.

```

## Binary Search Algorithm (iteratively)

```python
#Sorted List
def BinarySearch(a, key):
    low=0
    high=len(a)-1
    while low<=high:
        mid = (low+high)//2
        if key==a[mid]:
            return True
        elif key<a[mid]:
            high=mid-1
        else:
            low=mid+1
            return False
a=[15, 21, 47, 84, 96]
found=BinarySearch(a, 47)
print(found)
```

## Binary Search Algorithm (recursively)

```python
def BinarySearch(a, key, low, high):
    if high >= low:
        mid = low + (high - low)//2
        if key==a[mid]:
            return mid
        elif key<a[mid]:
            return BinarySearch(a, key, low, mid-1)
        else:
            return BinarySearch(a, key, mid+1, high)
    else:
        return "not present"
a=[15, 21, 47, 84, 96]
key=int(input())
print(BinarySearch(a, key, 0, len(a)-1))
```

## Selection Sort

```python
# Minimum Position
def SelectionSort(List):
    for i in range (len(List)):
        MinPos=i
        for j in range(i+1, len(List)):
            if List [MinPos]>List[j]:
                MinPos=j
            List[i], List[MinPos]=List[MinPos], List[i]            
    return List
List=[5, 2, 1, 9, 0, 4,6]
print(SelectionSort(List))
```

```python
# Counting sort in Python programming

def countingSort(array):
    size = len(array)
    output = [0] * size
    count = [0] * 10

    for i in range(0, size):
        count[array[i]] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    i = size - 1
    while i >= 0:
        output[count[array[i]] - 1] = array[i]
        count[array[i]] -= 1
        i -= 1

    for i in range(0, size):
        array[i] = output[i]


data = [4, 2, 2, 8, 3, 3, 1]
countingSort(data)
print("Sorted Array in Ascending Order: ")
print(data)
```

## Insertion Sort

```python
def InsertionSort(a):
    for i in range(len(a)):
        value=a[i]
        pos=i-1
        while pos>=0 and a[pos] > value:
            a[pos]=a[pos+1]
            pos=pos-1
            a[pos+1]=value
data=[15, 21, 47, 84, 96]
found=InsertionSort(data)
print(data)
```

## Bubble Sort

```python
def BubbleSort(a):
    for i in range (len(a)-1, 0, -1):
        for j in range (i):
            if a[j] > a[j+1]:
                a[j], a[j+1]=a[j+1], a[j]
data=[15, 21, 47, 84, 74,  96]
found=BubbleSort(data)
print(data)
```

## Quick Sort

```python
def QuickSort(a, low, high):
    if low < high:
        p = partition(a, low, high)
        QuickSort(a, low, p-1)
        QuickSort(a, p+1, high)
        
def partition(a, low, high):
    i=low-1
    pivot=a[high]
    for j in range(low, high):
        if a[j]<=pivot:
            i=i+1
        a[i], a[j]=a[j], a[i]
        
    a[i+1], a[high]= a[high], a[i+1]
    return i+1
a=[84, 21, 96, 15, 47]
QuickSort(a, 0, len(a)-1)
print(a)
```

```python
"""
Reverse a singly linked list. For example:
1 --> 2 --> 3 --> 4
After reverse:
4 --> 3 --> 2 --> 1
"""
#
# Iterative solution
# T(n)- O(n)
#
def reverse_list(head):
    """
    :type head: ListNode
    :rtype: ListNode
    """
    if not head or not head.next:
        return head
    prev = None
    while head:
        current = head
        head = head.next
        current.next = prev
        prev = current
    return prev


#
# Recursive solution
# T(n)- O(n)
#
def reverse_list_recursive(head):
    """
    :type head: ListNode
    :rtype: ListNode
    """
    if head is None or head.next is None:
        return head
    p = head.next
    head.next = None
    revrest = reverse_list_recursive(p)
    p.next = head
    return revrest
```

# Reverse a Linked List

```python


# Node class
class Node:

    # Init method to initialize the node object
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:

    # Init method to initialize head
    def __init__(self):
        self.head = None

    # Reverse method to reverse the linked list
    def reverse(self):
        prev = None
        current = self.head
        while (current is not None):
            next = current.next
            current.next = prev
            prev = current
            current = next
        self.head = prev

    # Push to insert a new node at the beginning
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    # Utility function to print the linked LinkedList
    def display(self):
        temp = self.head
        while (temp):
            print(temp.data)
            temp=temp.next


llist = LinkedList()
llist.push(20)
llist.push(40)
llist.push(60)
llist.push(80)

print("Given Linked List")
llist.display()
llist.reverse()
print("\nReversed Linked List")
llist.display()
```

### Circular Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data 
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None 
        # Function to insert in front(prepend)
    def prepend(self, data):
        new_node = Node(data)
        temp = self.head 
        new_node.next = self.head

        if not self.head:
            new_node.next = new_node
        else:
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
        self.head = new_node
        # Function to insert at the end(append)
    def append(self, data):
        if not self.head:
            self.head = Node(data)
            self.head.next = self.head
        else:
            new_node = Node(data)
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head
        # Function to print
    def print_list(self):
        temp = self.head 

        while temp:
            print(temp.data, end=" ")
            temp = temp.next
            if temp == self.head:
                break
        # Function to delete
    def remove(self, key):
        if self.head.data == key:
            temp = self.head 
            while temp.next != self.head:
                temp = temp.next 
            temp.next = self.head.next
            self.head = self.head.next
        else:
            temp = self.head 
            prev = None 
            while temp.next != self.head:
                prev = temp 
                temp = temp.next
                if temp.data == key:
                    prev.next = temp.next 
                    temp = temp.next


cllist = CircularLinkedList()
cllist.append("A")
cllist.append("B")
cllist.append("C")
cllist.append("D")

cllist.remove("A")
cllist.remove("C")
cllist.print_list()
```

### Doubly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        # Function to insert in end(append)
    def append(self, data):
        if self.head is None:
            new_node = Node(data)
            new_node.prev = None
            self.head = new_node
        else:
            new_node = Node(data)
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp
            new_node.next = None
        # Function to insert in front(prepend)
    def prepend(self, data):
        if self.head is None:
            new_node = Node(data)
            new_node.prev = None 
            self.head = new_node
        else:
            new_node = Node(data)
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
            new_node.prev = None
        # Function to print
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
    def add_after_node(self, key, data):
        temp = self.head
        while temp:
            if temp.next is None and temp.data == key:
                self.append(data)
                return 
            elif temp.data == key:
                new_node = Node(data)
                nxt = temp.next
                temp.next = new_node
                new_node.next = nxt
                new_node.prev = temp
                nxt.prev = new_node
            temp = temp.next

    def add_before_node(self, key, data):
        temp = self.head
        while temp:
            if temp.prev is None and temp.data == key:
                self.prepend(data)
                return
            elif temp.data == key:
                new_node = Node(data)
                prev = temp.prev
                prev.next = new_node
                temp.prev = new_node
                new_node.next = temp
                new_node.prev = prev
            temp = temp.next
        # Function to delete
    def delete(self, key):
        temp = self.head
        while temp:
            if temp.data == key and temp == self.head:
                # Case 1:
                if not temp.next:
                    temp = None 
                    self.head = None
                    return

                # Case 2:
                else:
                    nxt = temp.next
                    temp.next = None 
                    nxt.prev = None
                    temp = None
                    self.head = nxt
                    return 

            elif temp.data == key:
                # Case 3:
                if temp.next:
                    nxt = temp.next 
                    prev = temp.prev
                    prev.next = nxt
                    nxt.prev = prev
                    temp.next = None 
                    temp.prev = None
                    temp = None
                    return

                # Case 4:
                else:
                    prev = temp.prev 
                    prev.next = None 
                    temp.prev = None 
                    temp = None 
                    return 
            temp = temp.next


dllist = DoublyLinkedList()
dllist.prepend(0)
dllist.append(1)
dllist.append(2)
dllist.append(3)
dllist.append(4)
dllist.prepend(5)
dllist.delete(3)
dllist.add_after_node(2, 5)

dllist.print_list()
```

# Reverse a doubly linked list using recursion

```python
# Node of double Linked list
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None

# getnode method for new node
def getNode(data):
	new_node = Node(data)
	new_node.data = data
	new_node.next = new_node.prev = None
	return new_node

# push to insert node at the beginning of the Doubly Linked List
def push(head_ref, new_node):

	# since we are adding at the beginning, prev is always None
	new_node.prev = None

	# link the old list off the new node
	new_node.next = head_ref

	# change prev of head node to new node
	if (head_ref != None):
		head_ref.prev = new_node

	# move the head to point to the new node
	head_ref = new_node
	return head_ref

# Reverse method to reverse a doubly linked list
def Reverse(node):

	# If empty list, return
	if not node:
		return None

	# Otherwise, swap the next and prev
	temp = node.next
	node.next = node.prev
	node.prev = temp

	# If the prev is now None, the list has been fully reversed
	if not node.prev:
		return node

	# Otherwise, keep going
	return Reverse(node.prev)

# Function to print nodes in a given doubly linked list
def display(head):
	while (head != None) :
		print(head.data, end = " ")
		head = head.next

# Start with the empty list
head = None

# Create doubly linked:
head = push(head, getNode(10))
head = push(head, getNode(20))
head = push(head, getNode(30))
head = push(head, getNode(40))
print("Original list: ", end = "")
display(head)

# Reverse doubly linked list
head = Reverse(head)
print("\nReversed list: ", end = "")
display(head)


```

```python


from queue import Queue


class Stack:

    def __init__(self):

        # Taken two inbuilt queues
        self.q1 = Queue()
        self.q2 = Queue()

        # To maintain current number of elements
        self.curr_size = 0

    def push(self, x):
        self.curr_size += 1

        # Push x first in empty q2
        self.q2.put(x)

        # Push all the remaining elements in q1 to q2.
        while (not self.q1.empty()):
            self.q2.put(self.q1.queue[0])
            self.q1.get()

        # swap the names of two queues
        self.q = self.q1
        self.q1 = self.q2
        self.q2 = self.q

    def pop(self):

        # if no elements are there in q1
        if (self.q1.empty()):
            return
        self.q1.get()
        self.curr_size -= 1

    def top(self):
        if (self.q1.empty()):
            return -1
        return self.q1.queue[0]

    def size(self):
        return self.curr_size




s = Stack()
s.push(1)
s.push(2)
s.push(3)

print("current size: ", s.size())
print(s.top())
s.pop()
print(s.top())
s.pop()
print(s.top())
print("current size: ", s.size())
```

# Depth First Search

```python
from collections import defaultdict


class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        # If a key does not exist then it makes a default value of data type mentioned in the defaultdict
        
      # Two - Way Relation  
    def addEdge_U(self,ver_a,ver_b):
        self.graph[ver_a].append(ver_b)
        self.graph[ver_b].append(ver_a)

        # One - Way Relation
    def addEdge_D(self,ver_a,ver_b):
        self.graph[ver_a].append(ver_b)
        
        #depth first search
    def dfs_rec(self,start):
        # For every vertex you visit mark it as visited
        self.visited[start] = True
        # print that vertex
        print(start, end=" ")
        # now go its list of neighbours. self.graph[start] is all the neighbours of start.
        for i in self.graph[start]:
          # Waiting for the first unvisited neighbour
          if self.visited[i] == False:
            self.dfs_rec(i)
            # Exhaust all the possibilities for that neighbour.

    def dfs_runner(self,start):
        # Size of visited array is same as the number of vertices because each vertex has to be marked visited/unvisited
        self.visited = [False]*len(self.graph)
        # We have asked the user from which vertex to begin traversal.
        # Initially all vertices are unvisited
        # Now begin the traversal
        self.dfs_rec(start)
        
g=Graph()
#g.addEdge_D("0",'1')
#g.addEdge_D("0",'2')
#g.addEdge_D("1",'2')
#g.addEdge_D("1",'3')

g.addEdge_U(0,1)
g.addEdge_U(0,2)
g.addEdge_U(1,3)
g.addEdge_U(1,4)
g.addEdge_U(2,4)
g.addEdge_U(3,4)
g.addEdge_U(3,5)
g.addEdge_U(4,5)

g.dfs_runner(0)
```
